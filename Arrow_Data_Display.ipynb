```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, output_notebook, show
from bokeh.models.tools import HoverTool
```

```python
file_name = input('Enter file name')
file = pd.read_csv(file_name+'.csv', sep=',', header=12, skipinitialspace=True)
```

```python
data = pd.DataFrame (file)
```

```python
output_notebook() 
p = figure(title = file_name, x_axis_label = 'Frequency (Hz)', y_axis_label = 'Intensity')
p.line(data['frequency'], data['intensity'])
show(p)
```
