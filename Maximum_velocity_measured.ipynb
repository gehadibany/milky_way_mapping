```python

import pandas as pd
from bokeh.plotting import figure, output_notebook, show
from bokeh.models.tools import HoverTool
```

```python
filename = input('Enter file name:')+'.csv'
data = pd.read_csv (filename, header = 12, skipinitialspace = True)
```

```python
data
```

```python
output_notebook() 
p = figure(title = "velocity graph", x_axis_label = 'Velocity', y_axis_label = 'intensity', plot_width=800, plot_height=300)
p.line(data['corrected velocity'], data['intensity'])
p.add_tools(HoverTool(mode='vline'))
show(p)

import matplotlib.pyplot as plt
plt.figure (figsize=(8,3))
plt.plot (data['corrected velocity'], data['intensity'])
```
