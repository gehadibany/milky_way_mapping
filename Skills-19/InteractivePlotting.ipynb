This notebook how to generate interactive plots using Matplotlib.

Typically, we have used the directive `%matplotlib inline` to instruct Matplotlib to display plots within a notebook.  Using this system, plots are displayed immediately below the cell that creates them.  

Matplotlib also has the ability to disply plots in a separate window, providing interactivity such as zooming and panning, similar to that available using Bokeh.  This is achieved using the directive `%matplotlib qt` instead of `%matplotlib inline`.

The examples in this notebook plot a _sinc_ function.  This is defined as:

$$\text{sinc}(x) = \frac{\sin(x)}{x}$$

This function has the characteristic that it oscillates rapidly in the vicinity of the origin, but the amplitude dies away rapidly away from the origin.  

## 1 Inline plots

This first cell creates a plot of the sinc function and displays it inline in the usual way:  

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#Set the size of subsequent Matplotlib plots
plt.rcParams["figure.figsize"] = [12, 7.5]

# Data for plotting
t = np.arange(-5*np.pi, +5*np.pi, 0.01)
s = np.sinc(2 * np.pi * t)

# Create the plot
plt.plot(t, s)
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.title("Sinc function")
plt.grid()
plt.gca().add_patch(plt.Rectangle((-4, -0.22), 8, 1.22 , 
                                  linewidth = 2, linestyle = "--", 
                                  color = "red", fill = False))

plt.show()
```

## 2 Interactive plots

The plot in the previous cell is fine, but most of the detail is contained in a small region centred on the origin.  This is highlighted with the red dotted outline on the plot.

To study this part of the curve in more detail, it would be useful to be able to zoom in, and to make measurements of values at different points on the curve.  This can be done using the `%matplotlib qt` directive.  

### 2.1 Create interactive plot with qt
The following cell illustrates this.  When run, the plot will be generated in a separate window.  Controls in this window allow you to zoom and pan the plot, and to make changes to features such as the line colour and style.

**Important:** Depending on how your computer is set up, the window containing the plot may open behind the notebook. Look for an icon in the toolbar, or move the notebook window to one side to find it.

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib qt

#Set the size of subsequent Matplotlib plots
plt.rcParams["figure.figsize"] = [12, 7.5]

# Data for plotting
t = np.arange(-5*np.pi, +5*np.pi, 0.01)
s = np.sinc(2 * np.pi * t)

# Create the plot
plt.plot(t, s)
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.title("Sinc function")
plt.grid()

plt.show()
```

### 2.2 Compatibility

The `%matplotlib qt` directive can sometimes conflict with other software.  

If the example above does not run on your system, you can continue to create interactive plots using Bokeh instead.
