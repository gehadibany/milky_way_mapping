```python
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
from bokeh.models.tools import HoverTool
from IPython.display import display, clear_output
from bokeh.models import ColumnDataSource
from bokeh.models import Whisker
```

```python
def read_ARROW_data(filename):
    number_header_lines = 1
    dat = None
    with open(filename) as f:
        line = f.readline()
        while line[0] == '#' or line[0] == ',' or line[0].isspace():
            continue
            line = f.readline()
        dat = pd.read_csv(filename, header = number_header_lines, skipinitialspace = True)

    return dat
```

```python
spectrum_df = read_ARROW_data('Archive_Spectra.csv')

# Display the first few lines - does it look reasonable?
x= spectrum_df.iloc [1,:]
spectra = spectrum_df.drop(spectrum_df.index[0])

spectra.head(4)
```

```python
spec_no = len(spectra.columns)-1

#Cycle through this number of columns columns starting at column 1 (remember index 
#starts at 0 - which is the velocity column)
for idx in range(1,spec_no+1):
    # Get the name of the column
    colname=spectra.columns[idx]
    p1 = figure(title = "Spectral data from spectum column"+colname, 
          x_axis_label='Velocity (kms^-1)', 
          y_axis_label='Intensity',plot_width=1200, plot_height=450)
    p1.line(spectra['km per sec'],spectra[colname], color ='red')
    p1.add_tools(HoverTool(mode='vline'))
    
    src = ColumnDataSource(data=dict(
    y = spectra[colname],
    lower = spectra['km per sec'] - 5,
    upper = spectra['km per sec'] + 5))

    w = Whisker(base='y', 
          lower='lower',
          upper='upper', 
          line_color='white', 
          dimension='width', 
          source=src)

    p1.add_layout(w)
    show(p1)
    # Wait between plots - we'll ask the user whether they want to continue
    showNext = input('Show another plot? (y/n)')
    # For simplicity abort if anything other than "y" is entered.
    if showNext != 'y': 
        break
    # Clear the display before starting again - otherwise we get multiple plots
    clear_output(wait=True)
```
